# Importar bibliotecas necesarias


```python
import pandas as pd
import numpy as np
```

# Cargar el dataset

```python
df = pd.read_csv('C:\\Users\\Benjamin Villacorta\\Documents\\Python\\CursoML\\archive\\bank_transactions_data_2.csv')  # Reemplaza con la ruta correcta
```

# Convertir fechas a tipo datetime

```python
df['TransactionDate'] = pd.to_datetime(df['TransactionDate'])
df['PreviousTransactionDate'] = pd.to_datetime(df['PreviousTransactionDate'])
```

# Crear nuevas características basadas en las fechas

```python
df['TransactionYear'] = df['TransactionDate'].dt.year
df['TransactionMonth'] = df['TransactionDate'].dt.month
df['TransactionDay'] = df['TransactionDate'].dt.day
```

# Calcular la diferencia de tiempo entre transacciones

```python
df['TimeSinceLastTransaction'] = (df['TransactionDate'] - df['PreviousTransactionDate']).dt.total_seconds() / 3600
```

# Normalizar columnas numéricas

```python

numeric_cols = ['TransactionAmount', 'TransactionDuration', 'LoginAttempts', 'AccountBalance', 'CustomerAge']
df[numeric_cols] = (df[numeric_cols] - df[numeric_cols].mean()) / df[numeric_cols].std()
```

# Codificar variables categóricas

```python

categorical_cols = ['TransactionType', 'Location', 'DeviceID', 'MerchantID', 'Channel', 'CustomerOccupation']
df = pd.get_dummies(df, columns=categorical_cols, drop_first=True)
```

# Eliminar columnas innecesarias

```python

df.drop(['TransactionID', 'AccountID', 'IP Address'], axis=1, inplace=True)
```

# Manejar valores nulos si existiesen

```python

df.fillna(df.mean(), inplace=True)
```

# Guardar el dataframe limpio

```python

df_out=df.copy()
```
